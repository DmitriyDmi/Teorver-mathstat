## Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
zp = [100,80,75,77,89,33,45,25,65,17,30,24,57,55,70,75,65,84,90,150]
```

```python
srarifm = 0
for i in zp:
    srarifm = srarifm + i
srarifm = srarifm / len(zp)
print('Среднее арифметическое:', srarifm)
```

```python
srsqrtotkl = 0
for i in zp:
    srsqrtotkl = srsqrtotkl + (i - srarifm) ** 2
srsqrtotkl = (srsqrtotkl / (len(zp) - 1)) ** 0.5 # У нас явно указано, что дана выборка, поэтому знаминатель n-1
print('Среднее квадратичное отклонение:', srsqrtotkl)
```

```python
disp_smesh = 0
for i in zp:
    disp_smesh = disp_smesh + (i - srarifm) ** 2
disp_smesh = disp_smesh / len(zp)
print('Смещенная дисперсия:', disp_smesh)
```

```python
disp = 0
for i in zp:
    disp = disp + (i - srarifm) ** 2
disp = disp / (len(zp) - 1)
print('Несмещенная дисперсия:', disp)
```

## В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

Нас устроят 3 варианта:

    1. из первого ящика достали 2 белых, из второго 1
    2. из первого 1 белый, из второго 2
    3. из первого 0 белых, из втогого 3
    
Сразу запишем, что в 1м 3 небелых, во втором 7 небелых

```python
from math import factorial
def soch(k, n):
    return factorial(n) / (factorial(k) * factorial(n - k))
```

```python
p_1 = (soch(2,5) * soch(0,3) / soch(2,8)) * (soch(1,5) * soch(3,7) / soch(4,12))
p_2 = (soch(1,5) * soch(1,3) / soch(2,8)) * (soch(2,5) * soch(2,7) / soch(4,12))
p_3 = (soch(0,5) * soch(2,3) / soch(2,8)) * (soch(3,5) * soch(1,7) / soch(4,12))
p_itog = p_1 + p_2 + p_3
print('Вероятность 3 белых мячей:', p_itog)
```

## На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

Мы можем предположить, что вероятность того, что стрелял каждый из спортсменов равна 1/3, тогда применим формулу Баесса

```python
p_full = (1/3 * 0.9 + 1/3 * 0.8 + 1/3 * 0.6)
p1 = 1/3 * 0.9 / p_full
p2 = 1/3 * 0.8 / p_full
p3 = 1/3 * 0.6 / p_full
print('Вероятность что стрелял первый:', p1)
print('Вероятность что стрелял второй:', p2)
print('Вероятность что стрелял третий:', p3)
# Даже можно было не вводить 1/3, так как она сократиться до p_i/(0.9+0.8+0.6)
```

```python
# В сумме должна быть еденица, проверим
p1 + p2 + p3
```

## В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

Вероятность выбрать студента из А или В равна 0.25, вероятность выбрать студента из С равна 0.5

Применим формулу Баесса

```python
p_full = (0.25*0.8 + 0.25*0.7 + 0.5*0.9)
p_a = 0.25*0.8 / p_full
p_b = 0.25*0.7 / p_full
p_c = 0.5*0.9 / p_full
print('Вероятность учебы в А:', p_a)
print('Вероятность учебы в B:', p_b)
print('Вероятность учебы в C:', p_c)
```

```python
# Проверка
p_a + p_b + p_c
```

## Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

###  все детали

События независимы, но происходят одновременно.

Значит просто перемножим.

```python
p_1 = 0.1 * 0.2 * 0.25
print('Вероятность выхода из строя всех деталей:', p_1)
```

### только две детали

Во втором случае надо взять формулу полной вероятности и обратной вероятности для каждой детали поочередно.

```python
p_2 = 0.1 * 0.2 * 0.75 + 0.1 * 0.8 * 0.25 + 0.9 * 0.2 * 0.25
print('Вероятность поломки двух деталей:', p_2)
```

### хотя бы одна деталь

Хотя бы одна выйдет из строя, значит 1, 2 или все 3

Значит находим когда все 3 не выходят из строя и вычтем из 1

```python
p_3 = 1 - 0.9 * 0.8 * 0.75
print('Вероятность выхода хотя бы 1 детали:', p_3)
```

#### от одной до двух деталей

В этом случае нам нужны все вероятности кроме тех, при которых сломались все 3 или не сломались все 3

```python
p_4 = 1 - 0.9 * 0.8 * 0.75 - 0.1 * 0.2 * 0.25
print('Вероятность поломки 1 или 2 деталей:', p_4)
```
