### Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16. Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочная средняя M = 80, а объем выборки n = 256.

```python
import numpy as np
```

```python
std1 = 16
M = 80
n = 256
z_95 = 1.96
```

```python
x1 = M - z_95 * std1 / (n ** 0.5)
x2 = M + z_95 * std1 / (n ** 0.5)
print('Доверительный интервал:', (x1, x2))
```

### В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные: 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1 Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

```python
k = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]
t_95_9 = 2.262
n = 10
```

```python
std2 = np.std(k, ddof=1)
std2
```

```python
M = np.mean(k)
M
```

```python
x1 = M - t_95_9 * std2 / (n ** 0.5)
x2 = M + t_95_9 * std2 / (n ** 0.5)
print('Матожидание генеральной совокупности будет в интервале:', (x1, x2))
```

### Рост дочерей 175, 167, 154, 174, 178, 148, 160, 167, 169, 170 Рост матерей  178, 165, 165, 173, 168, 155, 160, 164, 178, 175 Используя эти данные построить 95% доверительный интервал для разности среднего роста родителей и детей.

```python
a = [175, 167, 154, 174, 178, 148, 160, 167, 169, 170]
b = [178, 165, 165, 173, 168, 155, 160, 164, 178, 175]
n = len(a)
```

```python
# Найдем несмещенные дисперсии и общую дисперсию
D1 = np.var(a, ddof=1)
D2 = np.var(b, ddof=1)
D = D1 / 2 + D2 / 2
D1, D2, D
```

```python
# Найдем стандартную ошибку разности средних
SE = (2 * D / 10) ** 0.5
SE
```

Найдем t критерий для 18 степеней свободы и двустороннего уровня значимости 0.05



```python
t = 2.101
```

```python
# Найдем разность средних по выборкам
M1 = np.mean(a)
M2 = np.mean(b)
delta = M1 - M2
delta
```

Отрицательное значение говорит лишь о том какое выборочное среднее больше, какое меньше. Можно брать по модулю.

```python
# Найдем доверительный интервал
X1 = delta - t * SE
X2 = delta + t * SE
print('Доверительный интервал разности средних:', (X1, X2))
```

Так как 0 попадает в доверительный интервал, можно говорить о том, что стредние выборок не имеют статистически значимых различий. M(a) = M(b) в 95% доверительном интервале.
