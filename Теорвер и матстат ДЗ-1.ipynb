```python
import math
```

```python
def soch(k,n):
    return math.factorial(n)/(math.factorial(k)*math.factorial(n-k))
```

### 1. Из колоды в 52 карты извлекаются случайным образом 4 карты. 

a) Найти вероятность того, что все карты – крести. 

б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

#### a) Найти вероятность того, что все карты – крести.

Крестей всего 13 в колоде, значит, благоприятных исходов будет С(4/13) * С(0/39)

Всего вариантов будет С(4/52)

С(0/39) = 1

Вероятность будет равна С(4/13) / С(4/52)

```python
soch(4,13)/soch(4,52)
```

#### б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

Думаю нужно воспользоваться формулой полной вероятности и сложить варианты когда тузов 1,2,3 или 4

Для каждого случая определим число благоприятных событий:

    1 туз: С(1/4) * С(3/48), 
    2: С(2/4) * С(2/48), 
    3: С(3/4) * С(1/48) 
    4: С(4/4) * С(0/48)

перепишем так: 

    1 туз: 4 * С(3/48), 
    2: С(2/4) * С(2/48), 
    3: 48 * С(3/4), 
    4: 4

По формуле общей вероятности будет общий знаминатель С(4/52), тогда сложим все благоприятные варианты и разделим

```python
(4 * soch(3,48) + soch(2,4) * soch(2,48) + 48 * soch(3,4) + 4)/soch(4,52)
```

### 2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

Задача на первый взгляд простая 

Благоприятных исходов 1

Всего исходов С(3/10)

Вероятность будет 1/С(3/10)

```python
1/soch(3,10)
```

### 3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

Благоприятные исходы - это 3 окрашеные детали, тоесть С(3/9)

А всего исходов С(3/24)

Поделим одно на другое, получим вероятность

```python
soch(3,9)/soch(3,24)
```

### 4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

В этой задаче важно понимать, что число благоприятных исходов 2, а не 1 как кажется на первый взгляд

Тем более С(2/2) = 2, а С(0/98) = 1

А всего вариантов купить 2 билета С(2/100)

Вероятность будет 2/С(2/100)

```python
2/soch(2,100)
```
