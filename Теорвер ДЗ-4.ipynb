### Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.

```python
a = 800
b = 200
```

```python
print('Среднее:', (a + b) / 2)
print('Дисперсия:', (a - b) ** 2 / 12)
```

### О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

```python
# Можно
b = 0.5
D = 0.2
# D = (a - b) ** 2 / 12 ----> a = (D * 12) ** 0.5 + b
a = (D * 12) ** 0.5 + b
M = (a + b) / 2
print(f'Правая граница: {a}, Среднее: {M}')
```

```python
# Проверка
(a - b) ** 2 / 12
```

### Непрерывная случайная величина X распределена нормально и задана плотностью распределения f(x) = (1 / (4 * sqrt(2pi))) * exp((-(x+2)**2) / 32)

Найдите:

    а). M(X)
    б). D(X)
    в). std(X) (среднее квадратичное отклонение)

```python
# Тут вычислять особо нечего
std = 4
M = -2
D = std ** 2
print('M(X) =', M)
print('D(X) =', D)
print('std(X) =', std)
```

### Рост взрослого населения города X имеет нормальное распределение. Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

    а). больше 182 см
    б). больше 190 см
    в). от 166 см до 190 см
    г). от 166 см до 182 см
    д). от 158 см до 190 см
    е). не выше 150 см или не ниже 190 см
    ё). не выше 150 см или не ниже 198 см
    ж). ниже 166 см.


а). больше 182 см, 1 - z(1std) = 1 - 0.8413 = 0.1587

б). больше 190 см, 1 - z(2std) = 1 - 0.9772 = 0.0228

в). от 166 см до 190 см, z(2std) - z(-1std) = 0.9772 - 0.1587 = 0.8185

г). от 166 см до 182 см, z(1std) - z(-1std) = 0.8413 - 0.1587 = 0.6826

д). от 158 см до 190 см, z(2std) - z(-2std) = 0.9772 - 0.0228 = 0.9544

е). не выше 150 см или не ниже 190 см, 1 - z(2std) + z(-3std) = 1 - 0.9772 + (1-0.9972)/2 = 0.0228 + 0.0014 = 0.0242

ё). не выше 150 см или не ниже 198 см, 1 - z(3std) + z(-3std) = 1 - 0.9972 = 0.0028

ж). ниже 166 см., z(-1std) = 0.1587

### На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

```python
M = 178
X = 190
D = 25
std = D ** 0.5
itog = (X - M) / std
print(f'Отклонение составит {itog} сигм')
```
