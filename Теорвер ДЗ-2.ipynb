### 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

Используем формулу бернулли.

Создадим для насала функцию для расчета сочетаний

```python
from math import factorial
```

```python
def soch(k, n):
    return factorial(n) / (factorial(k) * factorial(n - k))
```

Зададим переменные

```python
p = 0.8
q = 1 - p
k = 85
n = 100
```

```python
itog = soch(k, n) * (p ** k) * (q ** (n - k))
print(f'Вероятность попасть {k} раз из {n} равна: {itog}')
```

Подозрительно мало, проверим при k [75,85]

```python
for k in range(75,86):
    itog = soch(k, n) * (p ** k) * (q ** (n - k))
    print(f'Вероятность попасть {k} раз из {n} равна: {itog}')
```

Ответ 0.048061793700746355 проверен.

### 2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

Вероятность события(перегорания) очень мала, а испытаний много.

Используем формулу Пуассона.

Зададим переменные:

```python
from math import e
e
```

```python
p = 0.0004
n = 5000
lambda_n = p * n
lamps = [0, 2]
```

```python
for m in lamps:
    itog = (lambda_n ** m) * (e ** (-lambda_n)) / factorial(m)
    print(f'Вероятность перегорания {m} лампочек из {n} в первый день равна: {itog}')
```

### 3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

Зададим переменные:

```python
n = 144
k = 70
p = 0.5
q = 0.5
```

```python
itog = soch(k, n) * (p ** k) * (q ** (n - k))
print(f'Вероятность выпадаения орла {k} раз из {n} равна: {itog}')
```

### 4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

Декомпозируем на 3 отдельные задачи

#### Какова вероятность того, что все мячи белые?

Чтобы найти полную вероятность нужно перемножить вероятность вытащить 2 белых из первой и 2 белых из второй корзины

Р_1 = С(2,7) * С(0,3) / С(2,10)

Р_2 = С(2,9) * С(0,2) / С(2,11)

P_itog = P_1 * P_2

Вспоминаем что С(0,n) = 1

```python
k_1 = 7
k_2 = 9
n_1 = 10
n_2 = 11
m = 2
```

```python
P_1 = soch(m, k_1) / soch(m, n_1)
P_2 = soch(m, k_2) / soch(m, n_2)
P_itog = P_1 * P_2
print('Вероятность вытащить по 2 белых шара равна:', P_itog)
```

#### Какова вероятность того, что ровно два мяча белые?

Другими слвами, нас устроят варианты или 2 белых из 1й корзины 0 из 2й, или 2 белих из 2й корзины 0 из 1й, или по одному из каждой корзины.

Используем полную вероятность незвисимых событий:

P_itog = P_1_1 * P_2_1 + P_1_2 * P_2_2 + P_1_3 * P_2_3

где P_n_m - это вероятность события в случае m в корзине n

```python
# Для наглядности вычислений и простоты проверки не стал сразу сокращать С(0,n)=1 и C(1,n)=n
# Для первой корзины:
P_1_1 = soch(m, k_1) / soch(m, n_1) # Вытащили 2 белых
P_1_2 = soch(0, k_1) * soch(m, n_1 - k_1) / soch(m, n_1) # Вытащили 0 белых
P_1_3 = soch(1, k_1) * soch(1, n_1 - k_1) / soch(m, n_1) # Вытащили 1 белый

# Для второй корзины:
P_2_1 = soch(0, k_2) * soch(m, n_2 - k_2) / soch(m, n_2) # Вытащили 0 белых
P_2_2 = soch(m, k_2) / soch(m, n_2) # Вытащили 2 белых
P_2_3 = soch(1, k_2) * soch(1, n_2 - k_2) / soch(m, n_2) # Вытащили 1 белый
```

```python
P_itog = P_1_1 * P_2_1 + P_1_2 * P_2_2 + P_1_3 * P_2_3
print('Вероятность вытащить ровно 2 белых мяча будет:', P_itog)
```

#### Какова вероятность того, что хотя бы один мяч белый?

Переформулируем: нас устроит любой исход, кроме того, в котором вытащили все черные мячи

Найдем вероятность вытащить все черные мячи и вычтем из 1

P_itog = 1 - P_poditog

Р_1 = С(0,7) * С(2,3) / С(2,10)

Р_2 = С(0,9) * С(2,2) / С(2,11)

P_poditog = P_1 * P_2

```python
Р_1 = soch(0,7) * soch(2,3) / soch(2,10)
Р_2 = soch(0,9) * soch(2,2) / soch(2,11)
P_poditog = P_1 * P_2
P_itog = 1 - P_poditog
print('Вероятность хотя бы одного белого мяча будет:', P_itog)
```
